# Zipf's Law Analysis of Khalid's Lyrics

In this project, we analyze the lyrics of 64 songs by Khalid to see if they follow **Zipf's Law**, which describes the distribution of word frequencies in natural language. According to Zipf's Law, the frequency of words should follow a pattern where the most frequent word occurs roughly twice as often as the second most frequent word, three times as often as the third most frequent word, and so on.

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords


nltk.download('stopwords')

df = pd.read_csv("Khalid Data.csv")


lyrics = df['Lyric'].dropna().str.lower()

def clean_text(text):
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = text.split()
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    return tokens

all_words = []
for line in lyrics:
    all_words.extend(clean_text(line))


word_counts = Counter(all_words)
most_common = word_counts.most_common()


most_common[:10]


```

## Zipf's Law Analysis

Now, we will plot the word frequencies in a log-log scale to check if they follow Zipf's Law.


```python

ranks = np.arange(1, len(most_common) + 1)
frequencies = np.array([freq for word, freq in most_common])

plt.figure(figsize=(10, 6))
plt.loglog(ranks, frequencies, marker=".")
plt.title("Zipf's Law in Khalid's Lyrics")
plt.xlabel("Rank (log)")
plt.ylabel("Frequency (log)")
plt.grid(True, which="both", linestyle="--")
plt.show()
```

## Top 20 Most Frequent Words

Let's visualize the top 20 most frequent words in Khalid's lyrics using a bar chart.


```python
# Plot top 20 words
top_20 = most_common[:20]
words, freqs = zip(*top_20)

plt.figure(figsize=(12, 6))
sns.barplot(x=list(words), y=list(freqs), palette='plasma')
plt.xticks(rotation=45)
plt.title("Top 20 Most Frequent Words in Khalid's Lyrics")
plt.ylabel("Frequency")
plt.xlabel("Word")
plt.show()
```
